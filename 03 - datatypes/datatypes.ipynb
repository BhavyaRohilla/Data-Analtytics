## 4. String (str)
- A combination or sequence of chracters (alphabets, numbers, special chracters).
- We can use single quote or double quotes to represent strings.

```python
myString1 = 'Hello'
print(myString1)
```

```python
myString2 = "Hello"
print(myString2)
```

```python
myString3 = '''Hello'''
print(myString3)
```

### Multiline strings can be denoted using triple quotes, ''' or """

```python
# int
a = 0
```

```python
# int - bool
bool(a)
```

```python
# int - str
str(a)
```
